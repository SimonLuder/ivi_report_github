## Brushing / Linking example with Bokeh visualization library

#### Import Libraries

```python
import numpy as np
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.layouts import gridplot
from bokeh.models import ColumnDataSource
```

#### Prepare data

```python
x = np.arange(-10,11,1)
y1 = x
y2 = x**2
y3 = x**3
source = ColumnDataSource(data= dict(x=x, y1=y1, y2=y2, y3=y3))
```

#### Create Visualization

```python
#Figure 1
linear = figure(tools='lasso_select,box_select', plot_width=300, plot_height=300)
linear.circle(x="x", y="y1", size=13, color="blue", source=source)
linear.title.text = "Linear Scale"
#Figure 2
square = figure(tools='lasso_select,box_select',plot_width=300, plot_height=300)
square.circle(x="x", y="y2", size=13, color="green", source=source)
square.title.text = "Square Scale"
#Figure 3
exponent = figure(tools='lasso_select,box_select',plot_width=300, plot_height=300)
exponent.circle(x="x", y="y3", size=13, color="red", source=source)
exponent.title.text = "Exponential Scale"
output_notebook()

plot = gridplot([[linear, square, exponent]])
show(plot)
```

There are three subplots for three functions: f(x1) = x,  f(x2) = x^2,  f(x3) = x^3. It is possible to select individual points in one subplot via lasso-, or box-select brush (Change on top right). The corresponding values in the other two plots are automatically also selected.
